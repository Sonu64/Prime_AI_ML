```python
import json
```

```python
# load the data
def loadData(filePath):
    with open(filePath, "r") as f:
        data = json.load(f)
    return data
```

```python
data = loadData("data.json")
# print(data)
```

```python
def cleanData(data):
    chart = {"one":1, "two":2, "three":3, "four":4, "five":5}
    cleanedData = []
    uniqueUsers = set()
    for customer in data:
        
        # Fixing Inconsistencies like Type Inconsistency in rating, String vs Int.
        rawRating = customer['rating']
        rawRating = str(rawRating).strip().lower()
        if rawRating in chart.keys():
            rawRating = int(chart[rawRating])
            customer['rating'] = rawRating
        else:
            rawRating = float(rawRating)
            customer['rating'] = rawRating
            
        # Fixing Missing values of ages
        if customer.get('age') == None:
            customer['age'] = None
             
        # Removing Duplicate Values
        if customer['name'] not in uniqueUsers:
            uniqueUsers.add(customer['name']) 
            # Don't use .update(), use .add() to add new values to dictionary
            cleanedData.append(customer)    
    return cleanedData
```

```python
data = cleanData(data)
print(type(data[0]['rating']))
```

```python
def getInsights(data):
    # print(data[4])
    totalCustomers = len(data)
    
    # Find average rating
    sum = 0
    for customer in data:
        sum = sum + customer['rating']
    average = sum / totalCustomers
    print(f"Average Rating is {average}.")
    
    # Find percentage of users with Poor rating (<3)
    customersWithPoorRating = 0
    for customer in data:
        if customer['rating'] < 3:
            customersWithPoorRating += 1
    poorRatingPercentage = (customersWithPoorRating / totalCustomers) * 100
    print(f"{int(poorRatingPercentage)}% Users gave Poor Ratings.")
    
```

```python
getInsights(data)
```

```python
# Simple Recommendation Feature
def getRecommendations(data):
    recommendations = []
    
    for customer in data:
        currentRecommendation = {}
        currentRecommendation['name'] = customer.get('name')
        if float(customer['rating']) < 3.5:
            currentRecommendation['recommendations'] = "Apple, Vivo, Microsoft" # Other brands
        else:
            currentRecommendation['recommendations'] = "Asus" # Current Brand
        recommendations.append(currentRecommendation)
    return recommendations
```

```python
print(getRecommendations(data))
```
